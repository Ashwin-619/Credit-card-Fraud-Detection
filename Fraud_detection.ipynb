# **Training a Logistic regresssion model**

Using models from cuML to train in GPU for faster computation.
GridSearch to select best hyperparameter.

```python
# Importing GridSearchCV from sci-kit learn & LogisticRegression from cuML
from sklearn.model_selection import GridSearchCV
from cuml.linear_model import LogisticRegression


LR=LogisticRegression(penalty='l1',max_iter=1000)
param_grid = {'C': [ 0.01,0.1, 1, 10, 100]}
grid_search = GridSearchCV(LR, param_grid, cv=5, n_jobs=-1)
```

```python
grid_search.fit(x_train, y_train) # fitting data to grid search
```

```python
result=pd.DataFrame(grid_search.cv_results_)
result[['param_C','mean_test_score','rank_test_score']] # Viewing necessary columns
```

No big diffrence in test scores for diffrent values of C .
 So, default value is selected  

```python
# Traing with best parameters
LR=LogisticRegression(penalty='l1',max_iter=1000,C=1)
LR.fit(x_train,y_train)
```

```python
pred_LR=LR.predict(x_test) # Predicting testing values
# Creating confusion matrix
cm_LR=confusion_matrix(y_test,pred_LR)
```

```python
# Creating & plotting confusion matrix
cm_LR=confusion_matrix(y_test,pred_LR)
sns.heatmap(cm_LR,annot=True,fmt='g',xticklabels=['Fraud(1)','Legit(0)'],yticklabels=['Fraud(1)','Legit(0)'],cbar=False)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
```

```python
# Getting classification report
report=classification_report(y_test,pred_LR)
print(report)
```

```python
# Getting cross validation score
from sklearn.model_selection import cross_val_score
score=cross_val_score(LR,x_train,y_train,cv=5,scoring='recall') # scoring method is recall
```

```python
score_df=pd.DataFrame(score)
score_df
```

Checking if it is possible to increase recall for positive(Fraud) class by changing threshold value without sacrificing too much on precision

```python
proba=LR.predict_proba(x_test) # Getting probabilities for test set
proba_df=pd.DataFrame(proba[:,1]) # Keeping probabilities for positive class & converting into dataframe
proba_df.head()
```

```python
# Combining Probabilities and actual value into one dataframe
proba_df=pd.concat([proba_df.reset_index(drop=True),y_test.reset_index(drop=True)],axis=1,ignore_index=True)
proba_df.head()
```

```python
proba_df.columns=['Fraud','Class'] # Renaming columns
```

```python
proba_df.head()
```

```python
# Creating function to get recall and precision for given threshold list
from sklearn.metrics import recall_score , precision_score
def cm_threshold(threshold):
  threshold=np.round(threshold,2)
  for threshold in threshold:
      pred=np.where(proba_df['Fraud']>=threshold,1,0)

      print(f'threshold -->{threshold}   recall --->{np.round(recall_score(proba_df['Class'],pred),5)}   precision --->{np.round(precision_score(proba_df['Class'],pred),5)}')
```

```python
# Creating threshold list
threshold=[i for i in np.arange(0.10,0.5,0.05)]
cm_threshold(threshold)
```
